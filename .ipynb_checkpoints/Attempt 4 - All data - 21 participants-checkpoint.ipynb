```python
# this is the notebook that is using the most updated and complete spreadsheet, with the additional datasets included... 

# Data manipulation
import numpy as np
import pandas as pd

# Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Preferences
pd.options.display.max_columns=None
pd.options.display.max_rows=None
```

```python
# Read in the raw data
raw_data = pd.read_excel('Simulated Caregiving - Easy v Demanding - ALL.xlsx', sheet_name=None)
data = raw_data.copy()

# Merge all the sheets into a single dataframe and add sheet name as the "session variable
# Can we talk though this code? 
all_data = pd.DataFrame()
for k in data.keys():
    temp_df = data.get(k)
    temp_df['session'] = k
    all_data = pd.concat([all_data, temp_df])

# Relabel the cols for easier handling
all_data.columns = ['dolls', 'items', 'condition', 'interval_number', 'percent_interval_w_care',\
                    'percent_interval_w_care_easy', 'percent_interval_w_care_demanding','session']
    
# Reorganize the cols logically
all_data = all_data[[
    'session', 'dolls', 'items', 'condition', 'interval_number', 'percent_interval_w_care',\
    'percent_interval_w_care_easy', 'percent_interval_w_care_demanding'
]].reset_index(drop=True)

all_data
```
